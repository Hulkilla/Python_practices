# Ejercicio 1

Crea un programa que pida introducir una direccin de email por teclado. El programa debe imprimir en consola si la direccin de email es correcta o no en funcin de si esta tiene el smbolo @. Si tiene una @ la direccin ser correcta. Si tiene ms de una o ninguna @ la direccin ser errnea. Si la @ est al comienzo de la direccin de email o al final, la direccin tambin ser errnea,

```python
contadorArroba = 0

MiEmail = input("Introduce tu direccin de email: ")
for i in MiEmail:
  if (i == "@"):
    contadorArroba += 1

if contadorArroba == 1 and MiEmail.endswith("@") == False and MiEmail.endswith("@", 0, 1) == False:
  print("Email vlido")
else:
  print("Email no vlido")
```
