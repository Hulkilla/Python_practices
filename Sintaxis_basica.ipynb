```python
# Función suma de un número indeterminado de parámetros en python
# que se asegure de que todos los datos introducidos sean int o float

def suma(*args):
    total = 0
    for arg in args:
        if isinstance(arg, (int, float)):
            total += arg
        else:
            return "Error: Todos los argumentos deben ser int o float."
    return total
# En la función suma, el parámetro *args permite que la función acepte un número indeterminado de argumentos.
# Luego, se itera sobre cada argumento y se verifica si es un int o un float.
# Si lo es, se suma al total. Si no lo es, se devuelve un mensaje de error.
```

```python
print(suma(1, 2, 3, 4, 5))  # Output: 15
print(suma(1, 2, 3, 4, "5"))  # Output: Error: Todos los argumentos deben ser int o float.
```

# Listas

```python
Lista_ejemplo = [1, "casa", 3.72, True]
print(Lista_ejemplo)
# En un array o lista en python no todos los elementos tienen que ser del mismo tipo.
# OJO: El primer elemento está en la posición 0
```

```python
miLista = ["Maria","Pepe","Marta","Antonio"]
print(miLista[0])
print(miLista[-1])
print(miLista[0:3]) #El primer indice se incluye, pero el ulitmo no.
```

```python
miLista.append("Juan")
print(miLista[:])
#Append añade un elemento al final
```

```python
miLista.insert(2,"Sandra")
print(miLista[:])
#Insert lo hace en la posición uqe le digamos
```

```python
miLista.extend(["Marina", "Francisco", "Ana"])
print(miLista[:])
#Extend añade varios elementos al final
```

```python
miLista.index("Marta")
#Index nos da la posición de lo que busquemos
#Si en la lista tiene varias cosas iguales y lo buscamos con index, va a salir únicamente la primera posición que encuentre esta función
```

```python
print("Pepe" in miLista)
print(4 in miLista)
```

```python
miLista.remove("Ana")
print(miLista[:])
#Remove nos elimina el elemento que digamos, pero sólo el primero que aparezca en el array
```

```python
miLista.pop()
print(miLista[:])
#Pop elimina el último elemento del array
```

```python
Lista3 = miLista + Lista_ejemplo
print(Lista3)
#Aquí se concatenan distintas listas
```

```python
Lista4 = miLista*3
print(Lista4)
#Aquí se repite la lista tantas veces como digamos
```

Las tuplas son listas que no se las pueden modificar. Son inmmutables.

Si que se puede buscar y extraer información en nuevas tuplas y comprobar si hay un elemento en ella.

La lista se pone entre corchetes y la tupla con paréntesis.

Las listas se pueden convertir en tuplas y viceversa.


```python
tupla_ejemplo = ("Hola", 4, 7.456)
```

```python
print(tupla_ejemplo[2])
```

```python
mitupla = "Juan",3,46.2,14,"mu"
#Esto se considera un empaquetado de tupla porque no lleva paréntesis durante la creación de la tupla
print(mitupla)
```

```python
mitupla = ("Juan",3,46.2,14,"mu")
nombre,tiempo,coste,edad,sonido = mitupla
#Asigna el valor de la tupla a un nombre asigado
print(nombre)
print(tiempo)
print(coste)
print(edad)
print(sonido)
#Esto se conoce como desempaquetado de la tupla
```
